```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/category_year-csv.csv')
```

```python
print(df.columns)
```

```python
display(df)
```

```python
print(df.shape[0]-1)
df = df[0:10]
```

```python
plt.figure(figsize = (10, 6))
plt.plot(df.Year, df['mental health'], label = 'Mental Health')
plt.plot(df.Year, df['suicide'], label = 'Suicide')
plt.plot(df.Year, df['police'], label = 'Police')
plt.plot(df.Year, df['product'], label = 'Product')
plt.legend()

plt.xlabel('Years (2013-2022)')
plt.ylabel('Number of PFDs related to the category')

plt.savefig('spike_in_2021.png')

plt.show()
```

```python
plt.figure(figsize = (10, 20))
for column in df.columns: 
  plt.plot(df.Year, df[column], label = str(column))
plt.legend()

plt.xlabel('Years (2013-2022)')
plt.ylabel('Number of PFDs related to the category')

plt.show()
```
